```python
import cv2
from ultralytics import YOLO
```

```python
model = YOLO('best.pt')
```

```python
Video = cv2.VideoCapture('parking_1920_1080.mp4')
spots_status = [None for i in range(0,396)]
ret = True
while ret:
    ret, frame = Video.read()
    count = 0
    results = model.predict(frame)

    for r in results:
        boxes = r.boxes
        for box in boxes:
            
            b = box.xyxy[0]
            c = box.cls  
            # get box coordinates in (left, top, right, bottom) format
            x1 = int(b[0])
            y1 = int(b[1])
            x2 = int(b[2])
            y2 = int(b[3])
            if c == 1:
                count += 1
                frame = cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
            else:
                frame = cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 0, 255), 2)
    cv2.rectangle(frame, (80,20), (550,80), (0,0,0), -1)
    cv2.putText(frame, 'Available spots: {} / {}'.format(count, 396),
                (100,60), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255,255), 2)
    cv2.namedWindow('Video', cv2.WINDOW_NORMAL)
    cv2.imshow('Video', frame)
    if cv2.waitKey(25) & 0xff == ord('q'):
        break


Video.release()
cv2.destroyAllWindows()
```
